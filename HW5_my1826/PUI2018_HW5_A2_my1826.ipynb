```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
test = '''
SELECT *
FROM citibike
'''
```

```python
def get_data(query):
    try:
        return pd.read_csv(StringIO.StringIO(queryCartoDB(query)), sep=',')
    except ValueError as v:
        print (str(v))
```

# Task 1

```python
#Sort data by start_station_id, tripduration
#Only checking trips with duration <= 3 hours 

query1 = '''
SELECT * FROM citibike
WHERE tripduration < 10800
ORDER BY start_station_id, tripduration ASC
LIMIT 10
'''
#Only show the top 10 records (aka head and tail in SQL) 
table = get_data(query1)
table
```

```python
#Only show the last 10 records (aka head and tail in SQL) 

query1 = '''
SELECT * FROM citibike
WHERE tripduration < 10800
ORDER BY start_station_id, tripduration DESC
LIMIT 10
'''
table = get_data(query1)
table
```

```python
#List all unique start_station_id values 
query2 = '''
SELECT DISTINCT start_station_id FROM citibike
ORDER BY start_station_id
'''

table = get_data(query2)
table
```

```python
#Aggregation functions:
#- Count the number of trips (aka wc -l in SQL)
#- Find the average/min/max trip duration


query3 = '''
SELECT count(*), min(tripduration), avg(tripduration), max(tripduration)
FROM citibike
'''

table = get_data(query3)
table
```

```python
# Selecting trips started on Feb-02-2015 only
q1 = '''
select * from citibike 
where date(starttime) = '2015-02-02'

'''

table = get_data(q1)
table
```

```python
# Selecting trips started on the weekend
q2 = '''
select *
From citibike
where extract(dow from starttime) in(0,6)
'''
table = get_data(q2)
table
```

```python
#average trip duration started on the weekend
q3 = '''
select avg(tripduration)
From citibike
where extract(dow from starttime) in(0,6)
'''

table = get_data(q3)
table
```

```python
#average trip duration started on the weekdays
q4 = '''
select avg(tripduration)
From citibike
where extract(dow from starttime) not in(0,6)
'''

table = get_data(q4)
table
```

# Task 3

```python
#Showing the list of start station locations 

q5 = '''
select start_station_id,start_station_longitude, start_station_latitude
from citibike'''

table = get_data(q5)
table
```

```python
# Showing the number of trips started per station
q6 = '''
select count(tripduration),start_station_id
from citibike
group by start_station_id
'''
table = get_data(q6)
table
```

```python
# for stations within 500m of Time Square!

q7 = '''
SELECT count(tripduration),start_station_id
FROM citibike
WHERE st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
GROUP BY start_station_id
'''

table = get_data(q7)
table
```

# Task 4

```python
#Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare
q8 = '''
SELECT avg(tripduration)as duration,start_station_name
FROM citibike
WHERE st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
and EXTRACT(dow from starttime) in (0,6)
GROUP BY start_station_name
ORDER BY duration DESC
LIMIT 1
'''

table = get_data(q8)
table
```

```python
# Create lines for trips started from stations within 500m of Times
# Squares and lasted less than 2 hours. The number of trips per each
# pair of stations are output as attributes of these lines. 

q9='''
select count(start_station_id) as num_trips, start_station_name, end_station_name
from citibike
where st_dwithin(
CDB_LatLng(start_station_latitude, start_station_longitude)::geography, CDB_LatLng(40.7577, -73.9857)::geography, 500)
and tripduration <= 7200
Group by start_station_name, end_station_name

'''

table = get_data(q9)
table
```
