```python
import pandas as pd

url = "https://data.cityofnewyork.us/api/views/kpav-sd4t/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)
```

```python
import os

puidata = os.getenv('PUIDATA')
puidata
```

```python
df.head(7)
```

```python
df_reduced = df.drop(df.keys()[0:10]|df.keys()[12:], axis = 1)
df_reduced.head(7)
```

```python
%pylab inline
df_reduced.plot(x = "Salary Range From", y = "Salary Range To", kind = "scatter", title = 'NYC Salary Range Distribution')
```

Figure 1:
This graph displays the distribution of salary range of jobs listed on NYC jobs. Some takeawys:
1) Most of the jobs have starting salary from 40000 to 100000.
2) Jobs with starting salary from 50000 to 80000 have largest salary range.
